# JSON exercise

Download the data from [**here**](https://drive.google.com/file/d/1DGaX5AVfYhmWeb15lI-MzUbSKTYSz9fQ/view?usp=sharing) and answer following questions:
1. Find the 10 countries with most projects
2. What are top 10 sectors with projects?
3. Find the top 10 major project themes (using column 'mjtheme_namecode')
4. In 3. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

```python
import pandas as pd
from pandas.io.json import json_normalize
import json
```

```python
with open('world_bank_projects.json') as f:
    world_json = json.load(f)
json_normalize(world_json)
```

## Question 1

```python
pd.DataFrame(world_json)['countryshortname'].value_counts()[:10]
```

## Question 2

```python
sectors = json_normalize(world_json, record_path = 'sector')
pd.DataFrame(sectors)['Name'].value_counts()[:10]
```

## Question 3

```python
themes = json_normalize(world_json, record_path = 'mjtheme_namecode')
themes['name'].value_counts()
```

## Question 4

```python
themes = json_normalize(world_json, record_path = 'mjtheme_namecode')
themes[(themes['code'] == '1') & (themes['name'] == '')] = 'Economic management'
themes[(themes['code'] == '2') & (themes['name'] == '')] = 'Public sector governance'
themes[(themes['code'] == '3') & (themes['name'] == '')] = 'Rule of law'
themes[(themes['code'] == '4') & (themes['name'] == '')] = 'Financial and private sector development'
themes[(themes['code'] == '5') & (themes['name'] == '')] = 'Trade and integration'
themes[(themes['code'] == '6') & (themes['name'] == '')] = 'Social protection and risk management'
themes[(themes['code'] == '7') & (themes['name'] == '')] = 'Social dev/gender/inclusion'
themes[(themes['code'] == '8') & (themes['name'] == '')] = 'Human development'
themes[(themes['code'] == '9') & (themes['name'] == '')] = 'Urban development'
themes[(themes['code'] == '10') & (themes['name'] == '')] = 'Rural development'
themes[(themes['code'] == '11') & (themes['name'] == '')] = 'Environment and natural resources management'
themes['name'].value_counts()
```
